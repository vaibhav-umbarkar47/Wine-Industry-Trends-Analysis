```python
import numpy as np
product_name = []
country= []
wine_type = []
short_description= []
rating = []
reviews = []
price_per_bottle = []
quantity = []


for i in range(1,40):
    url = f'https://www.laithwaites.com/wine?country=Italy&country=France&country=United%20States&country=Spain&country=Australia&country=Portugal&country=Argentina&country=South%20Africa&country=New%20Zealand&country=United%20Kingdom&country=Germany&country=Israel&country=Georgia&country=Greece&hitsPerPage=96&page={i}'


    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }
    response = requests.get(url, headers = headers)
    if response.status_code == 403:
        print("There's a 403 Error!")
        break
    
    html_code = response.text

    
    soup = BeautifulSoup(html_code, "html.parser")

    containers = soup.find_all("li", class_="ais-Hits-item")
    for c in containers:
        # Product name
        name = c.find('div',attrs = {'class':'titleDescSale'})
        if name is None:
            product_name.append(np.nan)
        else:
            product_name.append(name.text)
    
        # country
        countrys = c.find('span', attrs = {'class':'country-title'})
        if countrys is None:
            country.append(np.nan)
        else:
            country.append(countrys.text)
            
        # wine_type 
        wine_types = c.find('span', attrs = {'class':'grapeVariety-title'})
        if wine_types is None:
            wine_type.append(np.nan)
        else:
            wine_type.append(wine_types.text)
    
        # short_description
        short_descriptions = c.find('span', attrs = {'class':'web-description-content'})
        if short_descriptions is None:
            short_description.append(np.nan)
        else:
            short_description.append(short_descriptions.text)

        # rating
        ratings = c.find('span', attrs = {'class':'star-icons'})
        if ratings is None:
            rating.append(np.nan)
        else:
            rating.append(ratings.text)
    
        # reviews
        reviewses = c.find('span', attrs = {'class':'review-count'})
        if reviewses is None:
            reviews.append(np.nan)
        else:
            reviews.append(reviewses.text)
    
        # price_per_bottle
        price_per_bottles = c.find('div', attrs = {'class':'price-section-single-bottle'})
        if price_per_bottles is None:
            price_per_bottle.append(np.nan)
        else:
            price_per_bottle.append(price_per_bottles.text)
    
        # quantity
        quantities = c.find('div', attrs = {'class':'sku-price'})
        if quantities is None:
            quantity.append(np.nan)
        else:
            quantity.append(quantities.text)

   
```

```python
print("Product_name",len(product_name))
print("country",len(country))
print("wine_type",len(wine_type))
print("short_description",len(short_description))
print("rating",len(rating))
print("reviews",len(reviews))
print("price_per_bottle",len(price_per_bottle))
print("quantity",len(quantity))
```

```python
import pandas as pd
df1 = pd.DataFrame({"product_name":product_name,
                    "country":country,
                    "wine_type":wine_type,
                    "short_description":short_description,
                    "rating":rating,
                    "reviews":reviews,
                    "price_per_bottle":price_per_bottle,
                    "quantity":quantity})
```
