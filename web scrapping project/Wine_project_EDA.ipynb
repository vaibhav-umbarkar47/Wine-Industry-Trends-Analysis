```python
import pandas as pd
```

```python
df2 = pd.read_csv('wines_data_without_EDA.csv')
```

```python
df2
```

```python
# Exctracting the year
import re
import numpy as np

df2['year'] = df2['product_name'].apply(
    lambda x: re.search(r'\b(19|20)\d{2}\b', x).group() if re.search(r'\b(19|20)\d{2}\b', x) else np.nan
)

df2['product_name'] = df2['product_name'].apply(
    lambda x: re.sub(r'\b(19|20)\d{2}\b', '', x).strip()
)

df2
```

```python

df2['quantity'] = df2['quantity'].astype(str).str.strip()

df2['price'] = df2['quantity'].str.extract(r'\$\s*([\d\.]+)').astype(float)

df2.head()
```

```python
df2['quantity'] = df2['quantity'].str.replace(r'\s*-\s*\$\s*[\d\.]+', '', regex=True)
df2['quantity'] = df2['quantity'].str.replace('bottles', '', regex=False).str.strip()
df2.head()
```

```python
df2 = df2.drop(columns=['rating'])
```

```python
df2.head()
```

```python
df2.rename(columns={'wine_type': 'Grapes_type'}, inplace=True)
```

```python
df2.head()
```

```python
df2.rename(columns={'price_per_bottle': 'price($)','Grapes_type':'grapes_type','price':'total_amount'}, inplace=True)
```

```python
df2.head()
```

```python
df2.rename(columns={'total_amount': 'total_amount(dollar)'}, inplace=True)
```

```python
df2.rename(columns={'price($)': 'price(dollar)'}, inplace=True)
```

```python
df2.head()
```

```python
df2.head()
```

```python
df2['price(dollar)']
```

```python
df2['price(dollar)'] = df2['price(dollar)'].astype(str).str.extract(r'\$([\d\.]+)')
```

```python
df2.head()
```

```python
df2['reviews'] 
```

```python
df2['reviews'] = df2['reviews'].astype(str).str.extract(r'(\d+)').astype(int)
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2 = df2.dropna(subset=['reviews']).reset_index(drop=True)
df2.head()
```

```python
df2.info()
```

```python
df2 = df2.dropna(subset=['quantity']).reset_index(drop=True)
```

```python
df2 = df2.dropna(subset=['price(dollar)']).reset_index(drop=True)
```

```python
df2 = df2.dropna(subset=['year']).reset_index(drop=True)
df2.head()
```
